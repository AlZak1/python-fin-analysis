# Matplotlib Visualization

* Learn how to use Matplotlib
* Subplots
* Multiple plots
* Bar plots

```python
import pandas as pd
```

```python
data = pd.read_csv('BJ.csv', index_col=0, parse_dates=True)
```

```python
data.head()
```

```python
data.plot()
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
data.plot()
```

```python
data['Close'].plot()
```

```python
fig, ax = plt.subplots()
data['Close'].plot(ax=ax)
ax.set_title('BJ')
```

```python
fig, ax = plt.subplots(2, 2)
data['Open'].plot(ax=ax[0, 0], title='Open')
data['High'].plot(ax=ax[0, 1], title='High')
data['Low'].plot(ax=ax[1, 0], title='Low')
data['Close'].plot(ax=ax[1, 1], title='Close')
plt.tight_layout()
```

```python
ax
```

```python
fig, ax = plt.subplots()
data['Volume'].loc['2022-01-02':'2022-06-01'].plot.bar(ax=ax)
```

```python
fig, ax = plt.subplots()
data['Volume'].loc['2022-01-02':'2022-06-01'].plot.barh(ax=ax)
```
